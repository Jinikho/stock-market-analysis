```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned_stock_data.csv", parse_dates= ['Date'], index_col='Date')
df.dropna(inplace=True)
df.head()
```

```python

plt.figure(figsize=(14,7))
plt.plot(df.index, df['Adj Close'], label='Adj Close Price')
plt.title('OMX Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,7))
ax.hist(df['Daily Return'].dropna(), bins=80)
plt.title('Distribution of Daily Return')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Rolling std (30d)'] = df['Daily Return'].rolling(window=30).std()
plt.figure(figsize=(14,7))
plt.plot(df.index, df['Rolling std (30d)'], label='30-Day Rolling Volatility')
plt.title('30-Day Rolling Volatility of OMX Stock')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```
